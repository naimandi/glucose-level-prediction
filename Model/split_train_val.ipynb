```python
import os
import shutil
import glob
import numpy as np
import pandas as pd

```

```python
# Set a fixed random seed for reproducibility. 
# This ensures code produce the same results.
np.random.seed(42)

data_dir = "/glucose-prediction/RP_Images/"
train_dir = data_dir + 'train/'
val_dir = data_dir + 'val/'

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

participant_folders = [name for name in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, name))]

for participant_folder in participant_folders:
    
    img_files = glob.glob(os.path.join(data_dir, participant_folder, "*.png"))
    
    np.random.shuffle(img_files)
    
    split_index = int(len(img_files) * 0.7)
    train_files = img_files[:split_index]
    val_files = img_files[split_index:]

    # Move the training images to the train directory
    for file in train_files:
        shutil.move(file, os.path.join(train_dir, os.path.basename(file)))

    # Move the validation images to the validation directory
    for file in val_files:
        shutil.move(file, os.path.join(val_dir, os.path.basename(file)))
```

```python
# ONLY RUN TO RESTORE TRAIN AND VAL DATASETS BACK TO PARTICIPANT FOLDERS

import os
import shutil

DATA_DIR = "/glucose-prediction/RP_Images/"
TRAIN_DIR = os.path.join(DATA_DIR, "train")
VAL_DIR = os.path.join(DATA_DIR, "val")

# Restore files in the train set
for file in os.listdir(TRAIN_DIR):
    participant_number = file.split('_')[1].lstrip("0")
    original_path = os.path.join(TRAIN_DIR, file)
    destination_path = os.path.join(DATA_DIR, f"P{participant_number}", file)
    shutil.move(original_path, destination_path)

# Restore files in the validation set
for file in os.listdir(VAL_DIR):
    participant_number = file.split('_')[1].lstrip("0")
    original_path = os.path.join(VAL_DIR, file)
    destination_path = os.path.join(DATA_DIR, f"P{participant_number}", file)
    shutil.move(original_path, destination_path)
```
